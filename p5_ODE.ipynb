# ODE

## Ejercicio 1
$y'= -5y+5t^2+2t$
* $0 \le t \le 1$
* $y(t_0)=1/3$
* $N=10$



```python
%autoreload 2
from src import ODE_euler
```

## Comparación con la solución real
La solución real de la ecuación diferencial es:
$y(t)=t^2 +1/3e^{-5t}$

* Grafique la solución real y la aproximación obtenida con el método de Euler en una misma gráfica.



```python
import matplotlib.pyplot as plt
import numpy as np

def f(t, y):
    return -5 * y + 5 * t **2 + 2 * t

a = 0
b = 1
N = 10

y0 = 1/3

resultado = ODE_euler(a=a, b=b,f=f, y_t0=y0 ,N=N)
ys = resultado[0]
ts = resultado[1]
h = resultado[2]

print("y=",ys,"t=",ts,"h=",h)
t_real = np.linspace(a, b, 11)
y_real = (t_real) ** 2 + 1/3 * np.exp(-5 * t_real)

plt.figure(figsize=(8, 5))
plt.plot(ts, ys, marker='o', linestyle='--', label='Aproximación Euler', color='skyblue')
plt.plot(t_real, y_real, linestyle='-', label='Solución Real', color='blue')
plt.title('Gráfica de y vs t usando el Método de Euler', fontsize=14)
plt.xlabel('t', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=10)
plt.show()

```

# Errores
Calcule el eeror relativo para diferentes valores de N.

N = 10

```python
# Calcular error relativo

errores_relativos = [abs((ys - y_real) / y_real) for ys, y_real in zip(ys, y_real)]

for i, error in enumerate(errores_relativos):
    print(f"Punto {i}: Error relativo = {error:.2f}")

promedio_error_relativo = sum(e for i, e in enumerate(errores_relativos)) / len(errores_relativos)
print("Error relativo promedio",promedio_error_relativo)
```
